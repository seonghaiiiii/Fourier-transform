```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.fft import fft, fftfreq

N = 1000
xa = np.linspace(0, 1, N)
```

```python
# 정현파 생성
frequencies = [10, 20, 50, 70]
y1 = np.sin(2*np.pi*frequencies[0]*xa)
y2 = np.sin(2*np.pi*frequencies[1]*xa)
y3 = np.sin(2*np.pi*frequencies[2]*xa)
y4 = np.sin(2*np.pi*frequencies[3]*xa)

# 그래프 생성
fig, ax = plt.subplots(nrows=4, ncols=1, figsize=(10,10))
ax[0].plot(xa, y1)
ax[0].set_title('10Hz')
ax[1].plot(xa, y2)
ax[1].set_title('20Hz')
ax[2].plot(xa, y3)
ax[2].set_title('50Hz')
ax[3].plot(xa, y4)
ax[3].set_title('70Hz')
plt.tight_layout()
plt.show()
```

```python
# 시간 간격 T 정의
T = xa[1] - xa[0]

# 새로운 신호 생성
signal = y1 + y2 + y3 + y4

# 푸리에 변환
frequency = fftfreq(N,T)[:N//2]
amplitude = fft(signal)[:N//2]/N

# 그래프 생성
fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(10,5))
ax[0].plot(xa, signal)
ax[0].set_title('Signal')
ax[0].set_xlabel('Time (s)')
ax[0].set_ylabel('$sin(x)$')
ax[1].plot(frequency, np.abs(amplitude))
ax[1].set_title('Frequency Domain')
ax[1].set_xlabel('Frequency (Hz)')
ax[1].set_ylabel('Amplitude')
ax[1].set_xlim([0,100])
plt.tight_layout(h_pad=4)
plt.show()
```

```python
t_n = 1
N = 1000
T = t_n/N

xb = np.linspace(0, t_n/4, num=int(N/4))

frequencies = [10, 20, 50, 70]
y1 = np.sin(2*np.pi*frequencies[0]*xb)
y2 = np.sin(2*np.pi*frequencies[1]*xb)
y3 = np.sin(2*np.pi*frequencies[2]*xb)
y4 = np.sin(2*np.pi*frequencies[3]*xb)

# 시그널 생성
signal = np.concatenate([y1, y2, y3, y4])

# 푸리에 변환
frequency = fftfreq(N, T)[:N//2]
amplitude = fft(signal)[0:N//2]

# 그래프 생성
fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(10,5))
ax[0].plot(xa, signal)
ax[0].set_title('Signal2')
ax[0].set_xlabel('Time (s)')
ax[0].set_ylabel('$sin(x)$')
ax[1].plot(frequency, np.abs(amplitude)/N)
ax[1].set_title('Frequency Domain')
ax[1].set_xlabel('Frequency (Hz)')
ax[1].set_ylabel('Amplitude')
ax[1].set_xlim([0,100])
plt.tight_layout(h_pad=4)
plt.show()
```
